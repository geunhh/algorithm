```python
#리스트 요소 탐색
def recur(idx, N, target) :
    #종료조건
    if idx==N : #못찾은 경우
        return -1
    
    #찾은 경우
    if nums[idx] == target :
        return 1
    
    return recur(idx+1, N, target)

nums = [10,20,55]
N = len(nums)
target = 99
recur(0, N, target)
```
