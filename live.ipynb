```python
# 주사위.
path = [] 

# 주사위 몇 개 던졌는지, 합이 몇인지.
def recur(level, total):
    # 가지치기
    if total >10:
        return
    
    # 기저조건 : 3개를 던졌을 때 종료.
    if level == 3:
        if total <= 10:
            print(path)            
        return
    
    # 후보군 탐색
    for i in range(1,7):
        path.append(i)
        recur(level+1, total+i)
        path.pop()

recur(0, 0)
```

```python
arr = ['A','B','C']
n= len(arr)


def get_sub(tar):
    for i in range(n):
        if tar & 0x1:    # 110 & 1, 11& 1, 1 & 1 --> B와 C만 나옴. 
            print(arr[i], end='')
        tar >>=1
get_sub(6) # 110
```

```python
# 친구와 카페 방문.
friends= ['a','b','c','d','e']
n = len(friends)

def get_count(tar):
    cnt =0
    for i in range(n):
        if tar & 0x1:
            # print(arr[i],end='')
            cnt+=1
        tar >>=1
    return cnt
    

result =0
for tar in range(0, 1<<n):
    if get_count(tar) >= 2:
        result+=1
print(result)
```

```python
arr =['A','B','C','D','E']
path = []

n = 3 # 몇 개 뽑을겨?
def run(lev,start):
    if lev == n:
        print(path)
        return
    
    for i in range(start,5):        # 처음 run 함수의 start 값은 0이다. 따라서 0~5까지 반복하며 재귀호출.
                                    # 만약 i가 3이 선택되는 경우 start 는 4가 됨.
        path.append(arr[i])
        run(lev+1,i+1)
        path.pop()
run(0,0)
```
