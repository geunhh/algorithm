```python
#리스트 메서드 직접 구현

#(1) len()
def get_len(input_list) :
    cnt = 0
    for _ in input_list :
        cnt += 1
    return cnt
print(get_len([1,2,3,3])) #4
 
#(2) count()
def get_count(input_list, find_value) :
    cnt=0
    for i in input_list:
       if find_value == i :
           cnt+=1
    return cnt

print(get_count([1,2,3,3], 3)) #2


#리스트 관련 함수 직접 구현
#(2) min(), max()
def get_min_max(input_list) :
    min_num = input_list[0]
    max_num = input_list[0]
    for num in input_list:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num= num
    return max_num, min_num
    

print(get_min_max([2,3,4,11,0])) #(0,11)

#(3) sum()
def get_sum(input_list) :
    num_sum =0
    for i in input_list:
        num_sum+=i
    return num_sum
    
#(3) avg() : 평균 구하는 함수
def get_avg(input_list) :
    num_sum = get_sum(input_list)
    
    return num_sum / len(input_list)

print(get_avg([1,2,3,4,5]))


#(4) std() : 표준편차 구하는 함수 (편차 제곱의 평균의 루트)
def get_std(input_list) :
    avg = get_avg(input_list)
    devi = []
    for i in input_list:
        devi.append((i-avg)**2)
    
    return get_sum(devi)/len(devi)

print(get_std([1,2,3,4,5]))

#(5) median() : 중앙값 구하는 함수
#오름차순 정렬 후 
#데이터의 개수가 홀수일 때 : 인덱스(데이터의 개수 // 2)의 값
#ex) 1,2,3,4,5 => 3
#데이터의 개수가 짝수일 때 : (인덱스(데이터의 개수 // 2 -1)의 값 + 인덱스(데이터의 개수 // 2 )의 값) / 2
#ex) 1,2,3,4 => (2+3)/2
def get_median(input_list) :
    input_list.sort()
    if len(input_list) % 2 ==1 :
        return input_list[len(input_list)//2]
    else:
        return (input_list[(len(input_list)//2)-1] + input_list[len(input_list)//2])/2
        
    
print(get_median([1,2,3,4])) #2.5
print(get_median([1,2,3,4,5])) #3
```

```python
#문자열 조작 메서드 직접 구현
#(1-1) find() : find_value는 문자 한개를 입력으로 일치하는 인덱스 반환, 없으면 -1 반환  
def get_find_ch(text, find_value) :    
    if find_value not in text:
        return -1
    else:
        return text.index(find_value)
print(get_find_ch('hello', 'l')) #2

#(1-2) find() : find_value는 문자열 입력으로 일치하는 인덱스 반환, 없으면 -1 반환  
# start로 지정된 인덱스 값 이후에 일치하는 문자열을 찾아서 인덱스 반환해야함
def get_find_str(text, find_str, start=3) :
    if find_str not in text[start:]:
        return -1
    else:
        return text.index(find_str)
print(get_find_str('hello', 'llo')) #2

#(2-1) split() : sep는 문자 한개를 입력으로 구분자대로 잘라서 리스트로 반환
def str_split(text, sep=' ') :
    new_lst=[]
    num=0
    for i in range(len(text)):
        if text[i] == sep:
            new_lst.append(text[num:i])
            num=i
    new_lst.append(text[num:len(text)])
    return new_lst
        

print(str_split('hello my name is gildong',  ' '))  #['hello', 'my' ,'name', 'is', 'gildong']

#(2-2) split() : sep는 문자열을 입력으로 구분자대로 잘라서 리스트로 반환
def str_split(text, sep) : 
    new_lst=[]
    sep_num=len(sep)
    num=0
    for i in range(len(text)):
        if text[i:i+sep_num] == sep:
            new_lst.append(text[num:i])
            num=i+1
    new_lst.append(text[num+sep_num-1:len(text)-1])
    return new_lst
    

print(str_split('hello my name is  gildong  gggeee',  '  ')) #['hello my name is', 'gildong ']

#(3) replace() : text에서 old문자열을 찾아서 new로 변경
def str_replace(text, old, new) :
    
    word_len = len(old)
    for i in range(len(text)):
        if text[i:i+word_len] == old :
            new_lst = text[:i] + new + text[i+word_len:]
    return new_lst

    
print(str_replace('hello my name is gildong', 'gildong', 'hani'))  #'hello my name is hani'

#(4) strip() : text에서 앞뒤로 remove_ch와 일치하는 문자를 제거
# remove_ch는 문자 한개를 입력으로 함
def str_strip(text, remove_ch) : 
    new_word =[]
    check=True
    for i in range(len(text)):
        if remove_ch != text[i]:
            start_num=i
            break
    j=0
    while check :
        if text[j] == remove_ch:
            j-=1
        else :
            end_num = j+1
            break
    new_word = text[start_num:end_num]
    return new_word

            



    
print(str_strip('----hello--', '-')) #hello

#첫글자가 영어의 소문자일 경우 대문자로 변경
#인터넷에 아스키 코드 참조하여 이용
#ord(문자한개) : 아스키코드값으로 반환
#chr(아스키코드) : 해당 아스키코드값을 문자로 반환
#(5) title()
def str_title(text) :
    if 'a'<= text[0] <= 'z':
        new_text = text.replace(text[0], chr(ord(text[0])-32) )
    return new_text
print(str_title('hello')) #Hello
```
