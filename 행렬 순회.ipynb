# 이차원 리스트로 표현한 행렬 순회

### 행 우선 순회


```python
#각 행별로 인덱스를 통한 순차적으로 접근
#%%
#              0열,1열,2열
num_matrix = [[10,20,30], #0행
              [30,50,60], #1행
              [80,45,59]] #2행

#각 행별로 인덱스를 통한 순차적으로 접근
for i in range(len(num_matrix)) :
    for j in range(len(num_matrix[0])) :
        print(num_matrix[i][j], end=' ')
    print()
```

```python
#각 행별로 값을 통한 순차적 접근, 행렬 값 수정 불가능
for row in num_matrix :
    for col in row :
        print(col, end=' ')
    print()
print('-'*50)
```

### 열 우선 순회


```python
#각 열별로 순차적으로 접근
#행의 값을 먼저 증가하여 처리
for j in range(len(num_matrix[0])) :
    for i in range(len(num_matrix)) :
        print(num_matrix[i][j], end=' ')
    print()
```

### 지그 재그 순회


```python
#홀수 행은 열의 마지막 인덱스에서 첫 인덱스로 접근 
print("원본:")
print(num_matrix)
N, M = len(num_matrix), len(num_matrix[0])
for i in range(N) :
    for j in range(M) :
        print(num_matrix[i][j + (M-1-2*j) * (i%2)], end=' ')
    print()
```

### 대각선 순회 또는 윗삼각, 아랫삼각 순회


```python
arr = [[1,2,3],
       [4,5,6],
       [7,8,9]]
N, M = len(num_matrix), len(num_matrix[0]) #행크기, 열크기

#오른쪽 위 삼각 순회
print('오른쪽 위 삼각 순회')
for i in range(N) :
    for j in range(M) :
        if i < j :
            print(arr[i][j], end=' ')
    print()

#왼쪽 아래 삼각 순회
print('왼쪽 아래 삼각 순회')
for i in range(N) :
    for j in range(M) :
        if i > j :
            print(arr[i][j], end=' ')
    print()

# 대각선(\) 순회
print('대각선(\) 순회')
for i in range(N) :
    for j in range(M) :
        if i == j :
            print(arr[i][j], end=' ')
    print()


# 대각선(/) 순회
print('대각선(/) 순회')
for i in range(N) :
    for j in range(M) :
        if N-1-i == j :
            print(arr[i][j], end=' ')
    print()
```

### 델타(방향 벡터)를 통한 순회
+ 2차 배열의 한 좌표에서 인접 배열 요소를 탐색하는 방법


```python
arr = [[1,2,3],
       [4,5,6],
       [7,8,9]]
N, M = len(num_matrix), len(num_matrix[0]) #행크기, 열크기

#          상,     하,     좌,       우
dirs = [[-1, 0], [1, 0], [0, -1], [0, 1]] #[[행 인덱스 증가값, 열 인덱스 증가값], ]

for i in range(N):
    for j in range(M):
        print(f"현재 좌표:{(i,j)}", end='==>인접좌표:')
        #현재 좌표에서 인접한 요소에 대한 처리
        for di, dj in dirs:
            ni, nj = i + di, j + dj
            if 0<=ni<N and 0<=nj<M:
                print(f"{(ni, nj)}", end=',')
        print()
```

### 전치 행렬 : 행과 열을 뒤바꾼 행렬


```python
from copy import deepcopy
#              0열,1열,2열
num_matrix = [[10,20,30], #0행
              [30,50,60], #1행
              [80,45,59]] #2행


N, M = len(num_matrix), len(num_matrix[0])

tr_m = deepcopy(num_matrix)
for i in range(N) :
    for j in range(M) :
        if i < j :
            tr_m[i][j], tr_m[j][i] = tr_m[j][i], tr_m[i][j]
print(tr_m)
print('-'*50)

#zip함수 활용 전치 행렬 만들기
tr_m = list(zip(*num_matrix))
print(tr_m)
#[(10, 30, 80), 
# (20, 50, 45), 
# (30, 60, 59)]
```
